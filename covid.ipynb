# **Estudando JOIN**

```python
# Configurando o index para o join globalmente para a coluna id
pessoas = pessoas.set_index('id')
covid = covid.set_index('id')
```

```python
# Criando um novo DataFrame com join da tabela pessoas com covid por um index temporário 'id pessoa'
df = pessoas.join(covid.set_index('id_pessoa'))
df.head(10)
```
